```python
import os
from skimage import io

repo_path = '/Users/anneitrheim/PycharmProjects/Colorize-Black-and-White-Photos/'
file_name = [f for f in os.listdir(repo_path + 'images-colored') if f.endswith(('.jpg', '.JPG'))]

# create a dictionary of the pictures (read in as arrays)
X_DICT_BW_TRAIN = dict()
Y_DICT_COLOR_TRAIN = dict()

for img_id in file_name:
    bw = io.imread(os.path.join(repo_path + 'images-bw', img_id))
    color = io.imread(os.path.join(repo_path + 'images-colored', img_id))
    X_DICT_BW_TRAIN[img_id] = bw
    Y_DICT_COLOR_TRAIN[img_id] = color

print("Dictionary of pictures created.")
```

```python
#plot random images from our dataset
import random
rand_img = random.randint(0,len(X_DICT_BW_TRAIN.keys()))
img = list(X_DICT_BW_TRAIN.keys())[rand_img]

import matplotlib.pyplot as plt
plt.subplot(1, 2, 1)
plt.imshow(X_DICT_BW_TRAIN[img], cmap='binary_r')
plt.subplot(1, 2, 2)
plt.imshow(Y_DICT_COLOR_TRAIN[img])
```

```python
heights = []
widths = []
for i in X_DICT_BW_TRAIN.keys():
    heights.append(X_DICT_BW_TRAIN[i].shape[0])
    widths.append(X_DICT_BW_TRAIN[i].shape[1])
    
import numpy as np
heights = np.array(heights)
widths = np.array(widths)

print("         min, mean, max")
print("heights:", heights.min(), round(heights.mean(),0), heights.max())
print("widths: ", widths.min(), round(widths.mean(),0), widths.max())
```
